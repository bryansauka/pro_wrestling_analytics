# Pro Wrestling Event Analysis (Since 2021)

## Overview
This notebook analyzes professional wrestling events for AEW and WWE from 2021 onward, focusing on event counts, average attendance, and broadcast characteristics. Consistent filters are applied throughout: only events from 2021 or later, and only those classified as TV-Show, Pay Per View, or Premium Live Event.

---

## Data Preparation

- **Source:** `report_pro_wrestling_events.csv`
- **Date Range:** 2021 and later
- **Event Types Included:** TV-Show, Pay Per View, Premium Live Event
- **Brands:** AEW, WWE

All analyses below use this filtered dataset.

---
